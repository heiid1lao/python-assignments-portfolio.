```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# === (a) First Partial Derivatives ===
x, y = sp.symbols('x y')
f = sp.exp(x + y) * sp.cos(y) + y**2

df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("∂f/∂x =", df_dx)
print("∂f/∂y =", df_dy)

# === (b) Gradient Vector and Magnitude at (1, -1) ===
g = x**2 * y + y**2
grad_g_x = sp.diff(g, x)
grad_g_y = sp.diff(g, y)

grad_at_point = (
    grad_g_x.subs({x: 1, y: -1}),
    grad_g_y.subs({x: 1, y: -1})
)

magnitude = sp.sqrt(grad_at_point[0]**2 + grad_at_point[1]**2)
print("Gradient at (1, -1):", grad_at_point)
print("Magnitude of gradient:", magnitude)

# === (c) Second Partial Derivatives ===
h = sp.log(x**2 + y**2 + 1)
d2h_dx2 = sp.diff(h, x, 2)
d2h_dy2 = sp.diff(h, y, 2)
d2h_dxdy = sp.diff(sp.diff(h, x), y)

print("∂²h/∂x² =", d2h_dx2)
print("∂²h/∂y² =", d2h_dy2)
print("∂²h/∂x∂y =", d2h_dxdy)

# === (d) Contour Plot of j(x, y) = x³ - 2xy + y³ ===
j = x**3 - 2*x*y + y**3
j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=60, cmap='plasma')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 2xy + y^3$')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# === (e) Custom Contour Plot of k(x, y) = e^{-x^2} * sin(y) ===
k = sp.exp(-x**2) * sp.sin(y)
k_func = sp.lambdify((x, y), k, 'numpy')
Z2 = k_func(X, Y)

plt.contour(X, Y, Z2, levels=60, cmap='coolwarm')
plt.colorbar()
plt.title('Contour plot of $k(x, y) = e^{-x^2} \cdot \sin(y)$')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
